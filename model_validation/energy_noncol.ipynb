```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd

file_list = [
    '/home/zhonghua/Filt_Event/recE/combined_1e8_1e9_M_params.csv',
    '/home/zhonghua/Filt_Event/recE/combined_1e9_1e10_M_params.csv',
    '/home/zhonghua/Filt_Event/recE/combined_1e10_1e11_M_params.csv',
    '/home/zhonghua/Filt_Event/recE/combined_1e11_1e12_M_params.csv',
]
# data_frames = [pd.read_csv(file) for file in file_list]
data=pd.concat((pd.read_csv(file) for file in file_list), ignore_index=True)
plt.hist(data['recE'], bins=100, alpha=0.7, label='Reconstructed Energy')
plt.show()
```

```python
recE_10_15 = data[(data['recE'] > 1.0) & (data['recE'] < 1.5)]
recE_15_20 = data[(data['recE'] > 1.5) & (data['recE'] < 2.0)]
recE_20_25 = data[(data['recE'] > 2.0) & (data['recE'] < 2.5)]

bins=50
plt.hist(recE_10_15['Eage'], bins=bins, alpha=0.7,density=True, label='recE: 1.0-1.5')
plt.hist(recE_15_20['Eage'], bins=bins, alpha=0.7,density=True, label='recE: 1.5-2.0')
plt.hist(recE_20_25['Eage'], bins=bins, alpha=0.7, density=True,label='recE: 2.0-2.5')
plt.xlabel('Eage')
plt.legend()
plt.show()
# R_mean
plt.hist(recE_10_15['R_mean'], bins=bins, alpha=0.7,density=True, label='recE: 1.0-1.5')
plt.hist(recE_15_20['R_mean'], bins=bins, alpha=0.7,density=True, label='recE: 1.5-2.0')
plt.hist(recE_20_25['R_mean'], bins=bins, alpha=0.7, density=True,label='recE: 2.0-2.5')
plt.xlabel('R_mean')
plt.legend()
plt.show()
```
